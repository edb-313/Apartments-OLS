The variables are: 

* Municipalities - name of the municipalities and territorial area
* waste total - total waste in the municipality	
* collected waste	- total selective collection waste in the municipality
* indifferent waste - total indifferent waste in the municipality
* unemployment - total unemployment  in the municipality
* tourism - tourism revenue per capita
* crime  - crime per capita
* wage  - wage per capita
* bank evaluation M2	
* tax - real estate tax	per capita
* purchasing power - purchasing power per capita
* population - total of the population	

Create new variables that may be adequate for the model, such as the selected waste percentage. 
Obviously, some variables are unnecessary, and some dates are virtually repeated.

1.	Import needed libraries, Import data
2.	Analyze variables
3.	Create a model that explains the price


```python
# import libraries
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
# read data
url = "https://github.com/masterfloss/data/raw/main/realEstate2018.xlsx"
df = pd.read_excel(url)
```

```python
# change data types
df = df.astype({'Municipalities':'string'})
df["waste total"] = pd.to_numeric(df["waste total"],errors = "coerce").astype(float)
df["bank evaluation M2"] = pd.to_numeric(df["bank evaluation M2"],errors = "coerce").astype(float)
df["collected waste"] = pd.to_numeric(df["collected waste"],errors = "coerce").astype(float)
df["indifferent waste"] = pd.to_numeric(df["indifferent waste"],errors = "coerce").astype(float)
df["unemployment"] = pd.to_numeric(df["unemployment"],errors = "coerce").astype(float)
df["tourism"] = pd.to_numeric(df["tourism"],errors = "coerce").astype(float)
df["wage"] = pd.to_numeric(df["wage"],errors = "coerce").astype(float)
df["population"] = pd.to_numeric(df["population"],errors = "coerce").astype(float)
df = df.set_index("Municipalities")
#create variables
df["selected waste percentage"] = df["collected waste"]/df["waste total"]*100
#clean data,
df= df.dropna()
df = df.drop(["waste total","collected waste", "indifferent waste"],axis=1)
df = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)] #could use to remove outliers in a specific column

#create x and y
x = df.drop(["bank evaluation M2"], axis = 1)
y = df["bank evaluation M2"]
```
